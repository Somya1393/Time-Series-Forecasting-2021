```python
from statsmodels.tsa.stattools import adfuller
```

```python

def check_stationary(x):
    adf_result = adfuller(x)
    if adf_result[1]<0.05:
        print('Series is stationary')
    else:
        print('Series is non-stationary')
```

```python
# we will try to make series stationary by 1st order differencing
```

```python
diff_1st_order_stationary = check_stationary(df_lang['en'].diff().dropna())
```
